```python
df['bmi'].fillna(0)
np.mean(df.bmi)
```

```python
df.fillna(np.mean(df.bmi))
```

```python
mode_value = df['bmi'].value_counts()
```

```python
df['bmi'].fillna(np.median(df.bmi))
```

```python
mode_bmi = df['bmi'].mode()
```
